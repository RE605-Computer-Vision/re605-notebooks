```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_detections(img_tensor, boxes, labels, scores, score_threshold=0.5):
    img = transforms.ToPILImage()(img_tensor)
    img = np.asarray(img)
    boxes = boxes.detach().numpy()
    labels = labels.detach().numpy()
    scores = scores.detach().numpy()
    n_boxes = labels.shape[0]
    for i in range(n_boxes):
        if scores[i] > score_threshold:
            x0 = int(boxes[i,0])
            y0 = int(boxes[i,1])
            x1 = int(boxes[i,2])
            y1 = int(boxes[i,3] )
            cv.rectangle(img, (x0, y0), (x1,y1), (0,0,255), 2)
    plt.imshow(img)
```

```python
show_detections(img_tensor, predictions[0]['boxes'], predictions[0]['labels'], predictions[0]['scores'], score_threshold=0.6)
```
